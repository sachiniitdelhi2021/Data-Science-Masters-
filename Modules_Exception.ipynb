```python
#  always remember that whenever we write code for every function and every line of code we have to implement exception handling and  
#  also for every code we have to peform 'logging'  for logging every activity and different types of logs(DEBUG,ERROR,WARNING,CRITICAL,INFO) 
```

```python
# if we don't  do exception handling than during execution of code if code fails at any line than code below that will not run 
#  without exception handling if any module(created by us) fails than rest all below modules will also fail or simlpy not work 
# but with exception handling if code gives error at any line than it wil not affect the execution of rest code or below modules that's why it is necessary  
```

```python
a=open('test.txt','r')
```

```python
# any line or part of code which is suspicious to give error should be put inside 'try' block and if that code is giving error than to handle thta
#  we put another block 'except' block 
```

```python
try:
  a=open('test.txt','r')
except:
    pass

```

```python
#  now it is not showing any error although error is there and rest all lines of code below this will be executed with the help of 'pass'
```

```python
# with the help of 'try' and 'except' block we cannot handle compile time error(syntax,indentation etc) we can only handle runtime errors 
```

```python
try:
  a=open('test.txt','r'
except:
    pass
```

```python
try:
  a=5/4
except ArithmeticError as e:
  print('there is problem with the code',e )
else:
  print('the code is working fine')    # this line  will show only if try block if having no error
  try:
     a=open('test.txt','r')
  except:
    print('there is issue with the internal try block') 
```

```python
#  in the 'except' module some errors like IOError(i/p o/p type error), arithmeticerror  etc, are already available to use and put in except block
#  when using ''try'' and ''except'' blocks like above  
```
