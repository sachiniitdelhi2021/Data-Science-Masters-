```python
import logging 

logging.basicConfig(filename='test3.log',level=logging.DEBUG,format='%(asctime)s %(levelname)s %(message)s')

# suppose we want to print what we log inside file during logging at the same time using 'streamhandler'
console_log=logging.StreamHandler()
# setting level for streamhandler
console_log.setLevel(logging.INFO)
# format for the level
format='%(asctime)s %(levelname)s %(message)s'
console_log.setFormatter(format) 

# create custom logger
logging.getLogger(' ').addHandler(console_log)
logging.info('this is my first code test for log' )

logger1=logging.getLogger('user1')
logger2=logging.getLogger('user2')
logger3=logging.getLogger('user3')
logger4=logging.getLogger('user4')

logger1.info('this is info.. for logger one')
logger2.debug('this is a debug for logger two')
logger2.info('this is info for logger two')
```
