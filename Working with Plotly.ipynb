```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import cufflinks as cf
```

```python
# this is for using 'plotly' in my local system  
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
```

```python
# let's create a dataframe 'df' using 'pandas' & and giving randomly generated data to it generated using 'numpy'
df=pd.DataFrame(np.random.rand(300,3),columns=['a','b','c'])
```

```python
df
```

```python
# let's plot the data of column 'a' and 'b' and for getting the relation b/w  data of both columns
df.iplot(x='a',y='b',mode='markers')
```

```python
#  the plot shows that thers is no relation b/w the data of both columns the data of both is kind of scattered all arround 
```

```python
#  let's import titanic dataset from seaborn and assign that dataset to a variable called 'titanic'
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
# now say i want to know the no. of males and no. of females who have survived and relation b/w them using graph and plotting both on same plot as  
titanic.iplot(x='sex',y='survived',kind='bar')
```

```python
# now plotting the plot for 'age' & 'fare'
#  we can also change the them of background and for this 'cufflinks' are what used as
cf.set_config_file(theme='solar')    # solar, polar are one of the themes   
titanic.iplot(x='fare',y='age',mode='markers',xTitle='fare',yTitle='age',title='fare vs age plot')
```

```python
# plotting the 'box' plot 
df.iplot(kind='box')
```

```python
'''with the help of 'box' plot we can get the maxi. ,min.,25% ,50% and 75% percentile and in the givn data what data points are
outliers 1) if any value in data goes beyond maxi.. value than it is called +ve outlier 2) if any value of data goes below mini.. value
it is a -ve outlier .'''

# maxi.. value= q3+1.5*IQR
# mini.. value=q1-1.5*IQR
#  IQR is inter quartile range (difference of q3 and q1)
```
