```python
# while loop operations 

initial_speed = 100
final_speed = 100
count=1
while initial_speed< final_speed:
  print('speed is this')
  count=count+1
  if count==10:
    break          # every time count value is increasing by 1 after printing 'speed is this' and that value is compared with value of count 
                   # which we decided as 10 and if value is <10 while loop keep on running again and again till count becomes 10 than inbuilt function
                   # 'break' will break the while loop and no more printing 'speed is this'  
else:
  print('initial_speed is not < than final_speed')
  print(initial_speed)


```

```python
n=7
for i in range(0,n):
  for j in range(i):
    print(' ',end='')
  for j in range(0,7-i):
    print('+ ',end='')
  print('\r')
```
