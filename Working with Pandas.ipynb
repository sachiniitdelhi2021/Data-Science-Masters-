```python
# There is something called 'merge' and 'join' operations in pandas we will what they do and how much these two are different 
```

```python
d1={'k':['k1','k2','k3'],'A':['A1','A2','A3'],'B':['B1','B2','B3']}
d2={'k':['k4','k2','k3'],'C':['C4','C5','C6'],'D':['D4','D5','D6']}
```

```python
#  let's create 2 new dataframe like
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
#  pandas 'merge' operation for merging any no of dataframes like
pd.merge(df1,df2,on='k')             # 'merge' offers 5 different opeartions like (left,right,outer,inner & cross)
#  by-default it performs 'inner' which only shows the data of only those rows which are same for selected column like this 
```

```python
#  in above operation with (on='K') what will happen only those rows for column 'k' which are having same value in both dataframe 
#  will be in result and rest all will not be there and for all remaining columns data for those rows will be displaced 
```

```python
# if i use (how='left')
pd.merge(df1,df2,on='k',how='left')  #for how='left' the frame which is in left 'df1' so from this frame it will consider all columns
                                     # and for all remainig columns in df2 which are not matching with columns of df1 will be added with 
                                # only  those rows values which are having same data for both frames in column 'k' and NaN for not matching values 
                                # will give nan in columns of df2 

```

```python
## if i use (how='right')
 
pd.merge(df1,df2,on='k',how='right')
```

```python
# # if i use (how='outer')
pd.merge(df1,df2,on='k',how='outer')
# 
```
