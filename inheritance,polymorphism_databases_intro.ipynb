```python
# polymorphism-->> if any operator (say +) behaves in diff.. ways depending upon differnet situations than it is called polymorphism
```

```python
def test(a,b):
  return a+b
```

```python
print(test(2,4))                         # in this case addition of two integer will happen 
print(test('sachin','dangi'))            # in this case concatination of two stings will happen 
print(test([1,2,3],[4,5,6]))                 # in this case two list will append
```

```python
class ineuron:
  def msg(self):
    print('this is a message from ineuron')

class xyz:
  def msg(self):
    print('this is a message for xyz')
```

```python
def test(notes):   # this is a separate function calling this and giving the object of any class as it's argument then 
  notes.msg()     # that object will be traced inside it's class and then test function will perform notes.msg() operation
```

```python
i=ineuron()
```

```python
x=xyz()
```

```python
#  now calling function tets()
test(i)
```

```python
# again calling function test() but this time giving object of another class as it's argument
test(x) 
```

```python
# the above example is of 'polymorphism'  
```
