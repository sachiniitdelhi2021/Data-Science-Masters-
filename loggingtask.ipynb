```python
# (Q1)---- create a logger in ur code 
import logging as lg 
lg.basicConfig(filename='test.log',level=lg.INFO)
```

```python
# defining a function

def test1(*args):
  lg.info('first call the function, i/p any no. of arguments,  then run the function, u will get sum of all i/ps')
  lg.info('test function will give you additin of input arguments')
  
  sum=0
  for i in args:
    lg.info(str(i))
    sum = sum + i
  return sum 
```

```python
test1(120,60,20)
```
