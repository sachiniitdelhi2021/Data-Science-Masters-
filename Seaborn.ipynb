```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#  let's import a dataset called 'tips' from seaborn as 
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
# let's plot a plot for 'tip' vs 'total_bill' using seaborn as 
sns.relplot(x='tip',y='total_bill',data=tips)
```

```python
# let's say in any column we have data of type say-- true, false or yes and no type data then we can plot a graph by selecting the smoke column than we will have plot in two different colors
sns.relplot(x='tip',y='total_bill',data=tips,hue='smoker') 
# this 'c' is for color ,any column i select for 'c' based on the no. of groupings of data in that column we will get plot in that differnet colors
```

```python
#  as you can see in above plot that we are getting the plot for 'tip' vs 'total_bill' but the colour is divided into 2 based on no. of grouping of data in that column given for 'hue'
```

```python
# 
sns.relplot(x='tip',y='total_bill',data=tips,hue='day') 
```

```python
#  the no. of days in day column based on that the above plot is in that many different colors ,originally ploted for 'tip' vs 'totao_bill'
```

```python
# but looking at above plot we can't say on any day out of total peoples how many were smokers and how many were not 
#  we can do this by using another parameter called ---- 'style' as 
sns.relplot(x='tip',y='total_bill',data=tips,hue='day',style='smoker') 
```

```python
#  now we can see the no. of smokers and non-smokers based on the shape of dot for each of the day 
```

```python
# say based on tips we want to know which person gave the highest tip so be can set this using diffenet sizes for that as 
sns.relplot(x='tip',y='total_bill',data=tips,size='tip') 
```

```python
#  you can see the different sizes of dots with which we can easily see which person paid the highest tip
```
