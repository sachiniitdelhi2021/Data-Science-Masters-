```python
l=[]
for i in range(10):            # i%2==0 means every value of i will be divided by 2 and if reminder is 0 then that value of i will be added inside list 
  if i%2==0:
    l.append(i)
```

```python
l
```

```python
#  same operation did above, now by list comprehenstion 
[i for i in range(10) if i%2==0]     
```

```python
l=[]     # again a empty list
for i in range(10):
  if i%2==0:
    l.append('CD')
  else:
    l.append('NCD')    # CD-- i is completely divisible
                          # NCD -- i is not completely divisible




```

```python
l
```

```python
#  same above operation using list comprehensition 
['CD'if i%2==0 else 'NCD' for i in range(10) ]
```

```python
# generating nested lists
l=[]
for i in range(3):
  l.append([])
  for j in range(4):
    l[i].append(j)
l
```

```python
#  same operation we can do using list comprehensition
[[j for j in range(4)] for i in range(3)]
```
