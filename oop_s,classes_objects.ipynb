```python
# (Q1)--->>>>  create a class with name 'data' with class variables 

class data():
  def __init__(self,file_name,file_type,date,size):
    self.file_name=file_name
    self.file_type=file_type
    self.date=date
    self.size=size

  def file_open(self):
    f=open('test.txt','w')
    f.write('this file is created under file_open function using oops concept')
    f.close()
  
  def file_read(self):
    a=open('test.txt','r')
    
    print(a.read())
    a.close()

  def file_append(self):
    a=open('test.txt','a')
    a.write('\nhere i m appending new data in this file using append ')
    a.close()
    a=open('test.txt','r')
    print(a.read())


```

```python
# (Q2)--- implement exception handling for each operation and do logging of each and every activity but by creating a seperate function when that function is called every thing will be logged 
import logging as lg
lg.basicConfig(filename='datalog.log',level=lg.error,format='%(asctime)s %(logged data is)s')

class data():
  def __init__(self,file_name,file_type,date,size):
    self.file_name=file_name
    self.file_type=file_type
    self.date=date
    self.size=size

  def file_open(self):
    try:
       f=open('test1.txt','w')
       f.write('this file is created under file_open function using oops concept')
       f.close() 
    except Exception as e:
      self.logging(e)
      print('file with this name alerady exist',e )
    
  def file_read(self):
    try:
       f=open('test1.txt','r')
       print('data inside file is\n',f.read())
       f.close() 
    except Exception as e:
      self.logging(e)
      print('no such file exist',e)
     
  def file_append(self):
    try:
       f=open('test1.txt','a')
       f.write('\nhere i m appending new data in this file using append ')
       f.close()
       f=open('test1.txt','r')
       print('this file contains\n',f.read())
       f.close()
    except Exception as e:
      self.logging(e)
      print('no such file exist',e)

  def logging(self,log):     # function created for logging error inside every other fun. of class and will be called inside any func. using- self.logging()
    lg.error(log)
   

```
