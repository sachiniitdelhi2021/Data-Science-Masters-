```python
def test4(n):           # there is diff.. b/w return and yield and it is that, here ''for loop'' will keep on running till last operation and  
  l=[]                         # then it will append all results and will give final output in one go not after every operation 
  for i in range(n):
    l.append(i**3)
  return l
```

```python
test4(30)
```

```python
def test5(n):             # but in case of ''yield'we will get result of each opeation just after it's completion and will not
  for i in range(n):      # not wait till completion of last operation and then show all results together in one go 
    yield i**3             # also if we use yield then it remembers our function so that we can use later anytime like
```

```python
test5(30)
```

```python
for i in test5(30):
  print(i)
```

```python
# generating fibonacci series 
def fbs(n):
  a=1
  b=1
  l=[]
  for i in range(n):
    l.append(a)
    a,b=b,b+a
  return l
```

```python
fbs(5)
```

```python
#  same operation but with generator function ''yield''
def fbs1(n):
  a=1
  b=1
  for i in range(n):
    yield a
    a,b=b,a+b 
```

```python
fbs1(5)
```

```python
for i in fbs1(5):
  print(i)
```
