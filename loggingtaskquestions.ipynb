```python
# (Q1)---- create a logger in ur code 
import logging as lg 
lg.basicConfig(filename='test.log',level=lg.info)

# defining a function

def test(*args):
  lg.info(str(args))
  sum=0
  for i in args:
    sum = sum + i
  return sum 
```

```python
test(120,60,20)
```
