```python
class test:
                # init is not always required untill we are not giving any data unlike above classes 
  def f(self):
    print('this is from class test')
```

```python
class test1:

  def f(self):
    print('this is from class test1')
```

```python
#  creating a child class 
class test2(test,test1):
  def a(self):
    t=test()     # t is the object of class test
    t.f()
    t1=test1()    # t1 is the object of class test1 
    t1.f()
t2=test2()    # t2 is the object of child class test2
t2.a()
```

```python
t2.f()     # when we inherite multiple parent classes than in such case the child class by-default calls the function of first parent class 
```

```python
# EX-1---single parent class 
# parent class
class student:
  name='sachin'
  ph_no='9752050888'
  college='iit delhi'

  def college_dtls(self):
    print('his college is',self.college)

# child class
class form(student):
  def __init__(self):
    self.course_dtls='masters in design'
  def details(self):
    print('{} is stuent of {}, doing his {}'.format(self.name,self.college,self.course_dtls))
```

```python
sachin=form()
```

```python
sachin.details()
```
