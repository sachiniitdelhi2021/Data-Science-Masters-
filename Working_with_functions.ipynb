```python
# Q1--- create fun.. which can take any no. of any datatype data and based on that it will return seperate lists based on datatype
def Q1(*args):
  l1=[]
  l2=[]
  l3=[]
  l4=[]
  l5=[]
  l6=[]
  
  for i in args:
    if type(i)==int:
      l1.append(i)
    
    if type(i)==str:
      l2.append(i)
      
    if type(i)==bool:
      l3.append(i)
      
    if type(i)==float:
      l4.append(i)
      
    if type(i)==list:
      l5=l5+i
      
    if type(i)==tuple:
      l6.append(i)
  return l1+l2+l3+l4+l5+l6 # if i want one single  list 
#  if i want seperate lists than use -- return l1,l2,l3,l4,l5,l6 


```

```python
Q1(12,'sachin',23,'dangi',76, 12.98,[1,2,4],17.98,[9,98,6],(1,2,3,4),(9,8,5,57),True,False,True)
```

```python
# Q2--- use args and kwargs both and concatenate all i/p data of diff... type and return on single list 
def Q2(*args,**kwargs):
  l=[]
  l1=[]
  for i in args:
    l.append(i)
  for j in kwargs:
    l1.append(j)
  return l+l1
```

```python
Q2(12,'sachin',12.98,True,[1,2,3],(1,2,3),{1,2,3},a=12,b=19,c='dangi')
```

```python
#  imp function 
def test13(*args):
  dt={}
  for i in args:
    if type(i) in dt:
     dt[type(i)].append(i)
    else:
      dt[type(i)]=[i]
  return dt
```

```python
test13(1,2,3,[5,6,8],'sachin',2+3j,True)
```
