```python
# importing all basic lib... 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a dataframe using pandas and giving random data inside it using numpy as
# for freezing the all data values use 'seed' as 
np.random.seed(20)
df=pd.DataFrame(np.random.randn(1000,1),index=pd.date_range('2022-05-26',periods=1000),columns=['data'])
```

```python
df
```

```python
# if i say plot the above dataframe 
df.plot()
```

```python
df
```

```python
#  creating new columns
df['data1']=df.data.cumsum()   #this is for giving the cumulative sum of values of column 'data' as u can see 
df['data2']=df['data'].apply(lambda x:x*3)  # generating a new column by multiplying 3 with values of values of column 'data' 
```

```python
df
```

```python
#  let's plot this above dataframe as
df.plot()
```

```python
#  the blue graph is lying under the green graph 
#  we can change the size of this plot also as--- df.plot(figsize=(10,10))
df.plot(figsize=(10,5))
```

```python
#  you can see the blue colour graph under the green colour graph 
```

```python
# let's use 'seaborn' inside seaborn there are many datasets which are pre-loaded inside seaborn as 
# loading 'iris' dataset for flowers inside seaborn 
df1=sns.load_dataset('iris')  # df1 is another dataframe inside we are loading this data 
```

```python
df1
```

```python
#ploting the above dataframe as
#  we can set figure size ,graph title 
gf=df1.plot(figsize=(10,10),title='iris dataset')
#  for setting the 'x'&'y' axis labels we have to store this above graph in variable as
gf.set_xlabel('values of indexes')
gf.set_ylabel('data values of all columns')
```

```python
#  you can see 4 differnet graph with differnt colours in plot representing the data of all 4 columns 
```

```python
# for getting the stastitacial details of each column we can use -- describe()
df1.describe()    # from this we can clearly get the max. , min. and other details also for each column as 
```

```python
# let's say i want to plot graph for data of any particular row 
df1.iloc[0][0:4]  # for index '0'& for data in columns with index 0 to 4 excluding '4' 
```

```python
#  for above data in series form, for row with index as '0' plotting a different graph as 'bar' chart
gf1=df1.iloc[0][0:4].plot(figsize=(5,5),title='bar chart',kind='bar')
gf1.set_xlabel('columns')
gf1.set_ylabel('data values')
```

```python
# we can see on histogram also how the graph of each column is overlaping using-- stacked=True
# also i can set  the width for each vertical pillar of the histogram using-- bin=any number say 100
df1.plot(kind='hist',figsize=(5,5),stacked=True,bins=100,title='histogram of iris')
```

```python
#  you can see the overlap of each columns data 
# we can also make this plot in horiziontal orientation as 
df1.plot(kind='hist',figsize=(5,5),stacked=True,bins=100,title='histogram of iris',orientation='horizontal')
```

```python
# we can get seperate histograms for data of each column as 
df1.hist(figsize=(6,6),color='r',alpha=0.9)
#  i can set any color i want just google html color and copy the hexadecimal code of any color paste it here 
#  alpha-- is for contrast of the graph i think ,for high contrast give high value as 0.9 here in this case 
```

```python
#  same graph in different color , just google html color and copy the color's hexadecimal code and paste it here 
df1.hist(figsize=(6,6),color='#d600d6',alpha=0.9)
# (#7fb6ae) this is the color code for this color found on 'html color code ' 
```

```python
# scatter plot --in this we have to specify which parameters data will be ploted on x-axis and which other on y-axis 
#  say i want to see how sepal_length and petal_length are having some relation or not  
df1.plot.scatter(x='sepal_length',y='petal_length')
```

```python
# if i want to see the relation of two data sets with other data sets values like ---petal_length and sepal_length on petal_width 
#  than in this case we introduce something called --
df1.plot.scatter(x='sepal_length',y='petal_length',c='petal_width')
```

```python
''' now u can see in above graph ,if i pick any bundle of data points  represented by dots having values for both sepal_length
& petal_length in some range are also having values for petal_width in some range '''
''' the seperate bar that u see in side of above graph is called 'color bar'''
```

```python
# with this color bar in above graph we want to see the dependency of sepal_length & petal_length on petal_width   
df1.plot.scatter(x='sepal_length',y='sepal_width')
```

```python
# for making the above graph look more informative and good we can use ---1)label- for giving name to these dots 2)size(s)- for controlling the size of these dots 
df1.plot.scatter(x='sepal_length',y='sepal_width',label='data points',s=100)
```

```python
#  we can also plot hexabin for above datapoints like 
#  similar to scatter plot we can plot hexabin , with gridsize (for controlling the size of these grids as)
df1.plot.hexbin(x='sepal_length',y='sepal_width',label='data points',gridsize=10)
```

```python
# in our dataframe 'df1' we have 4 columns with numerical data if i want to plot all possible plots for all columns than i can do this by using scatter matrix as
#  using the function(plotting.scatter_matrix()) of pandas as 
pd.plotting.scatter_matrix(df1,figsize=(10,10))
```

```python
#  the above is called 'scatter matrix' which is basically the plot for each column data values against each other
#  as you can see along the diagonal we are getting histogram but i don't want histogram if i want 'bell' curve along the diagonal than  i can use-- diagonal='kde'
#  'kde'-- is called kernal density estimate 
pd.plotting.scatter_matrix(df1,figsize=(10,10),diagonal='kde')
```
